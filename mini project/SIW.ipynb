```python
# Перевод валют в USD
currency_conversion = {
    'EUR': 1.22,
    'GBP': 1.37,
    'INR': 0.014,
    'CNY': 0.16,
    'JPY': 0.0097,
    'MXN': 0.050,
    'PLN': 0.26,
    'CAD': 0.79,
    'DKK': 0.16,
    'SGD': 0.74,
    'HUF': 0.0033,
    'HRK': 0.16,
    'AUD': 0.76,
    'SEK': 0.12,
    'NOK': 0.11,
    'RUB': 0.013,
    'NZD': 0.70,
    'HKD': 0.13,
    'THB': 0.032,
    'ZAR': 0.066,
    'BRL': 0.18,
    'ARS': 0.011,
    'TWD': 0.036,
    'TRY': 0.12,
    'CHF': 1.09,
    'PHP': 0.020,
    'EGP': 0.064,
    'IDR': 0.000071,
    'CLP': 0.00105
}

def convert_to_usd(row):
    if row['salary_currency'] == 'USD':
        return row['salary']
    else:
        return row['salary'] * currency_conversion.get(row['salary_currency'], 1.0)

# Применяем функцию convert_to_usd к DataFrame для добавления столбца 'salary_usd'
df['salary_usd'] = df.apply(convert_to_usd, axis=1)


```

```python
def calculate_stats(df, profession=None):
    # Если профессия не указана, просто считаем статистику по всем данным
    if profession is None:
        df_filtered = df.copy()
    else:
        # Фильтруем DataFrame по указанной профессии
        df_filtered = df[df['job_title'] == profession]

    # Извлечение уникальных рабочих лет
    years = df_filtered['work_year'].unique()

   
    stats = {'year': [], 'average_salary': [], 'max_salary': [], 'min_salary': []}

    
    for year in years:
        # Получить подмножество данных за текущий год
        df_year = df_filtered[df_filtered['work_year'] == year]

        
        avg_salary = df_year['salary_usd'].mean()
        max_salary = df_year['salary_usd'].max()
        min_salary = df_year['salary_usd'].min()

        
        stats['year'].append(year)
        stats['average_salary'].append(avg_salary)
        stats['max_salary'].append(max_salary)
        stats['min_salary'].append(min_salary)

    # Создайте DataFrame из словаря
    stats_df = pd.DataFrame(stats)

    return stats_df

# Пример вызова функции с указанием профессии
status_df_engineer = calculate_stats(df, profession='Data Engineer')
print(f"Data Engineer:{status_df_engineer.to_string()}")

status_df_science = calculate_stats(df, profession='Data Scientist')
print(f"Data Scientist: {status_df_science.to_string()}")

status_df_analyst = calculate_stats(df, profession = 'Data Analyst')
print(f"Data Analyst: {status_df_analyst.to_string()}")

# Если не указана профессия, будут рассчитаны статистики по всем профессиям
status_df_all = calculate_stats(df)
print(f"All profession: {status_df_all.to_string()}")
```

```python
def calculate_monthly_stats(df):
    # Конвертировать год работы в месяц
    df['month'] = pd.to_datetime(df['work_year'], format='%Y').dt.month
    
    # Извлечь уникальные месяцы
    months = df['month'].unique()
    
    # Инициализировать пустой словарь для хранения статистики
    stats = {'month': [], 'average_salary_usd': [], 'max_salary_usd': [], 'min_salary_usd': []}
    
    # Проитерироваться по уникальным месяцам
    for month in months:
        # Получить поднабор данных для текущего месяца
        df_month = df[df['month'] == month]
        
        # Вычислить статистику для текущего месяца
        avg_salary_usd = df_month['salary_usd'].mean() / 12
        max_salary_usd = df_month['salary_usd'].max() / 12 
        min_salary_usd = df_month['salary_usd'].min() / 12
        
        # Добавить статистику в словарь
        stats['month'].append(month)
        stats['average_salary_usd'].append(avg_salary_usd)
        stats['max_salary_usd'].append(max_salary_usd)
        stats['min_salary_usd'].append(min_salary_usd)
    
    # Создать DataFrame из словаря
    monthly_stats_df = pd.DataFrame(stats)
    
    return monthly_stats_df

# Вызов функции для вычисления статистики по месяцам
monthly_stats_df = calculate_monthly_stats(df)

# Display monthly stats
print(monthly_stats_df.to_string())
```

```python
def count_employees_per_profession(df):
    # Группировка по профессии и подсчет уникальных значений
    profession_counts = df['job_title'].value_counts()
    return profession_counts

# Вызов функции для подсчета числа сотрудников по профессиям
employee_counts = count_employees_per_profession(df)
print(employee_counts)
```

```python
def calculate_salary(df):

    max_salary_per_profession_country = df.groupby(['job_title', 'company_location'])['salary_usd'].max()
    
    
    avg_salary_per_country = max_salary_per_profession_country.groupby('company_location').mean()
    
    return avg_salary_per_country



salary1 = calculate_salary(df)
print(salary1)
```

```python
def find_highest_paid_profession(df):
    # Группировка по профессиям и расчет  зарплаты
    avg_salary_per_profession = df.groupby('job_title')['salary_usd'].max()
    
     # Найдите профессию с самой высокой средней зарплатой
    highest_paid_profession = avg_salary_per_profession.idxmax()
    highest_avg_salary = avg_salary_per_profession.max()
    
    return highest_paid_profession, highest_avg_salary


highest_paid_profession, highest_avg_salary = find_highest_paid_profession(df)

print(f"The profession '{highest_paid_profession}' receives the highest salary of ${highest_avg_salary:.2f} USD per year.")
```

```python
def find_min_paid_profession(df):

    min_salary_per_profession = df.groupby('job_title')['salary_usd'].min()
    

    min_paid_profession = min_salary_per_profession.idxmin()
    min_avg_salary = min_salary_per_profession.min()
    
    return min_paid_profession, min_avg_salary


min_paid_profession, min_avg_salary = find_min_paid_profession(df)

print(f"The profession '{min_paid_profession}' receives the minimum salary of ${min_avg_salary:.2f} USD per year.")
```

```python
def calculate_salary_level(df):
    senior_avg_salary = df[df['experience_level'] == 'SE']['salary_usd'].mean()
    middle_avg_salary = df[df['experience_level'] == 'MI']['salary_usd'].mean()
    junior_avg_salary = df[df['experience_level'] == 'EN']['salary_usd'].mean()
    director_avg_salary = df[df['experience_level'] == 'EX']['salary_usd'].mean()
    
    return {
        'senior': senior_avg_salary,
        'middle': middle_avg_salary,
        'junior': junior_avg_salary,
        'director': director_avg_salary
    }


salary_stats = calculate_salary_level(df)
print(salary_stats)
```

```python
import matplotlib.pyplot as plt
def plot_salary(df, profession=None):
    if profession is None:
        df_filtered = df
        title = 'Salary Distribution for All Professions'
    else:
        df_filtered = df[df['job_title'] == profession]
        title = f'Salary Distribution for {profession}'

    plt.figure(figsize=(10, 6))
    plt.hist(df_filtered['salary_usd'], bins=20, color='skyblue', edgecolor='black')
    plt.xlabel('Salary (USD)')
    plt.ylabel('Frequency')
    plt.title(title)
    plt.grid(True)
    plt.show()

# Plot salary distribution for all professions
plot_salary(df)
```
